```python
# import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Prepare our data

```python
# load input dataset
train_path = "./train.csv"
train = pd.read_csv(train_path, index_col="id")

train.rename(columns = {'loss':'trx_count'}, inplace = True)
print(train.shape)
train.head()
```

```python
# data distribution
train.describe()
```

```python
# information about dataset
train.info()
```

```python
# seperate features and target variable
X = train.drop("trx_count", axis=1)
y = train.trx_count
feature_names = X.columns.to_list()
```

```python
# data standardization
from sklearn.preprocessing import StandardScaler

standizer = StandardScaler()
X = standizer.fit_transform(X)
```

# Fit a Poisson Regression model to the data

```python
# model training
import statsmodels.api as sm

exog = sm.add_constant(X)
poission_model = sm.GLM(y, exog, family=sm.families.Poisson())
result = poission_model.fit()
```

```python
# model summary
result.summary()
```

# Make prediction on the test set

```python
# predict
test_path = "./test.csv"
test = pd.read_csv(test_path, index_col="id")
test.head()
```

```python
print(test.shape)
test.head()
```

```python
# predict for test data
from sklearn.model_selection import KFold

X_test = sm.add_constant(standizer.transform(test))
prediction_test = []

kfold = KFold(shuffle=True, random_state=42)
for train_idx, valid_idx in kfold.split(X, y):
    
    features = sm.add_constant(X[train_idx])
    
    model = sm.GLM(
        y[train_idx], 
        features, 
        family=sm.families.Poisson()
    )
    result = model.fit()
    
    prediction_test.append(result.predict(X_test))
```

```python
preds = np.column_stack(prediction_test).mean(axis=1)
```
